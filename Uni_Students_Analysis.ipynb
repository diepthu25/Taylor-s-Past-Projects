# Heat Map

```python
label_cols = ['DepressionLabel', 'AnxietyLabel', 'StressLabel']
value_cols = ['DepressionValue', 'AnxietyValue', 'StressValue']
features = df.drop(columns=value_cols+label_cols).columns
print(features)
```

```python
#limit 3 category (3 tables/category)
features_per_category = 3

#label title
category_map = {
    'DepressionLabel': 'Depression',
    'StressLabel': 'Stress',
    'AnxietyLabel': 'Anxiety'
}

#iterate & extract name
for label in label_cols:
    label_name = category_map[label]

    # select subset of a feature
    subset_features = features[:features_per_category]

    print(f'\n{label_name} Analysis\n')
    for col in subset_features:
        pivot_table = pd.crosstab(df[col], df[label]) #create pivot table
        pivot_table = pivot_table.div(pivot_table.sum(axis=1), axis=0) * 100

        plt.figure(figsize=(10, 8))
        sns.heatmap(pivot_table, annot=True, fmt='.2f', cmap='YlGnBu', cbar_kws={'label': '%'})
        plt.title(f'{label_name} vs {col}')
        plt.xlabel('')
        plt.ylabel('')
        plt.yticks(rotation=0)
        plt.xticks(rotation=45)  # rotate x-axis labels if necessary
        plt.tight_layout()  # adjust layout to fit labels
        plt.show()
```

# Prediction Model

```python
from sklearn.preprocessing import OrdinalEncoder

# tạo copy để ko trùng
df_pm = df.copy()

# tìm những column category
categorical_cols = df_pm.select_dtypes(include=['object']).columns

# sử dụng ordinalencoder vào các category
ordinal_encoder = OrdinalEncoder()
df_pm[categorical_cols] = ordinal_encoder.fit_transform(df_pm[categorical_cols])

# updated dataframe
df_pm.head()
```

```python
for label in label_cols:
    cols = list(features).copy()
    cols.append(label)
    data = df_pm[cols].corr()
    plt.figure(figsize=(20,8))
    sns.heatmap(data, annot=None, fmt='.3f', cmap='coolwarm')
    plt.title(f'{label} Correlation')
    plt.show()
```

```python
def distplot(feature, frame, color='r'):
    plt.figure(figsize=(8,3))
    plt.title("Distribution for {}".format(feature))
    ax = sns.distplot(frame[feature], color= color)

for feat in value_cols: distplot(feat, df)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

print('Model Accuracy')
model_accuracy = {}
for label in label_cols:
    X = df_pm.drop(columns=label_cols)
    y = df_pm[label]

    # split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # train model
    model = RandomForestClassifier(random_state=0, n_estimators=30, max_depth=8)
    model.fit(X_train, y_train)

    # prediction
    preds = model.predict(X_test)

    # tính accuracy
    accuracy = accuracy_score(y_test, preds) * 100
    model_accuracy[label] = accuracy

    # Print accuracy for each label
    print(f'{label}: {accuracy:.2f}%')
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier  #import library
from sklearn.metrics import accuracy_score

print('Model Accuracy')
model_accuracy = {}
for label in label_cols:
    X = df_pm.drop(columns=label_cols)
    y = df_pm[label]

    # split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    # train model
    model = DecisionTreeClassifier()
    model.fit(X_train, y_train)

    # prediction
    preds = model.predict(X_test)

    # calculate accuracy
    accuracy = accuracy_score(y_test, preds) * 100
    model_accuracy[label] = accuracy

    # Print accuracy for each label
    print(f'{label}: {accuracy:.2f}%')
```
